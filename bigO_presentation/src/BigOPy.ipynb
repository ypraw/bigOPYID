# BigO : Code Complexity as Data Grows

### Contoh 1 (Pangkat)

```python
from math import pow
import timeit
```

```python
def pangkat1(x,y):
    hasil=1
    for i in range(0,y):
        hasil= x*hasil
    return hasil
```

```python
def pangkat2(x,y):
    hasil=x**y
    return hasil
```

```python
def pangkat3(x,y):
    hasil=pow(x,y)
    return hasil
```

```python
print(f'with loop \t {timeit.timeit("pangkat1(100,100)",globals=globals(),number=100000)}')
print(f'with ** operator {timeit.timeit("pangkat2(100,100)",globals=globals(),number=100000)}')
print(f'with pow \t {timeit.timeit("pangkat3(100,100)",globals=globals(),number=100000)}')
```

## Find Minimum Value on the List

```python
import time
from random import randrange
```

```python
# O(N^2)
def findmin1(data):
    lowest = data[0]
    for i in data:
        isSmallest = True
        for j in data:
            if i > j:
                isSmallest = False
            if isSmallest:
                lowest = i
    return lowest
```

```python
# 
def findmin2(data):
    lowest = data[0]
    for i in data:
        if i < lowest:  # di eksekusi selama n
            lowest = i  # dieksekusi 1 kali
    return lowest  #   dieksekusi 1 kali
```

```python
def findmin3(data):
    return min(data)
```

### Result find Minimum 1

```python
x1=[]
y1=[]
for listSize1 in range(1000, 10001, 1000):
    example_list = [randrange(10000) for x in range(listSize1)]
    
    start_time = time.time()
    print(f'minimum value: {findmin1(example_list)}')
    end_time = time.time()
    et=end_time-start_time
    print(f'size list: {listSize1}, time execution: {et:5f}')
    x1.append(listSize1)
    y1.append(et)
print()
```

### Result Find Minimum 2

```python
x2=[]
y2=[]
for listSize2 in range(1000, 10001, 1000):
    example_list2 = [randrange(10000) for x in range(listSize2)]
    start_time = time.time()
    print(f'minimum value: {findmin2(example_list2)}')
    end_time = time.time()
    et2=end_time-start_time
    print(f'size list: {listSize2}, time execution: {et2:.5f}')
    x2.append(listSize2)
    y2.append(et2)
print()
```

```python
x3,y3=[],[]
for listSize3 in range(1000, 10001, 1000):
    example_list3 = [randrange(100000) for x in range(listSize3)]
    start_time = time.time()
    print(f'minimum value: {findmin3(example_list3)}')
    end_time = time.time()
    et3=end_time-start_time
    print(f'size list: {listSize3}, time execution: {end_time-start_time:.5f}')
    x3.append(listSize3)
    y3.append(et3)

```

# Searching Complexity

### Logarithmic Complexity

```python
def binarySearch(lst, search):
    lower_bound = 0
    upper_bound = len(lst) - 1
    while True:
        if upper_bound < lower_bound:
            print("Not found.")
            return -1

        i = (lower_bound + upper_bound) // 2

        if lst[i] < search:
            lower_bound = i + 1
        elif lst[i] > search:
            upper_bound = i - 1
        else:
            return [search, i]
    return -1


```

### Liniear Complexity

```python
def linearSearch(sortedList, target):
    for i in range(len(sortedList)):
        if (sortedList[i] == target):
            return [i, sortedList[i]]
    return -1


```

```python
""" Testing Runtime """
size1, size2, time1, time2 = [], [], [], []

for listSize in range(1000, 100001, 1000):
    exampleList = [x for x in range(listSize)]
    start_time = time.time()
    print(
        f"binary search O(log N) : {binarySearch(exampleList,exampleList[len(exampleList)-1])}"
    )
    end_time = time.time()
    et = end_time - start_time
    print(f'size list: {listSize}, time execution: {et:.5f}')
    size1.append(listSize)
    time1.append(et)

print(" ")

for listSize2 in range(1000, 100001, 1000):
    exampleList2 = [x for x in range(listSize2)]
    start_time2 = time.time()
    print(
        f"Linier Search O(N) : {linearSearch(exampleList2,exampleList2[len(exampleList2)-1])}"
    )
    end_time2 = time.time()
    et2 = end_time2 - start_time2
    print(f'size list: {listSize2}, time execution: {et2:.5f}')
    size2.append(listSize2)
    time2.append(et2)
```

### Visualization Sum of List

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(size1, time1, 'g-', alpha=1,label="Binary Search")
plt.grid(True)

plt.plot(size2, time2, 'r-', alpha=1,label="Linear Search")
# Label the axes and provide a title
plt.title("haha")
plt.legend()
plt.show()
```
